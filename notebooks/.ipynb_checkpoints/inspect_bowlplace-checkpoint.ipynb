# Bowlplace Dataset Inspection
Use this notebook to explore `datasets/bowlplace_human.hdf5` and `datasets/bowlplace_robot.hdf5`.
The helper utilities below will list the dataset structure, report tensor shapes, and preview image-like entries if present.

```python
from pathlib import Path
try:
    import h5py
except ImportError as exc:
    raise RuntimeError('Please install h5py first: pip install h5py') from exc
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (8, 6)
DATA_DIR = (Path.cwd().resolve() / '../datasets').resolve()
human_h5 = DATA_DIR / 'bowlplace_human.hdf5'
robot_h5 = DATA_DIR / 'bowlplace_robot.hdf5'
human_h5, robot_h5
```

```python
def is_image_dataset(dset: h5py.Dataset) -> bool:
    if dset.ndim < 3:
        return False
    if dset.shape[-1] not in (1, 3, 4):
        return False
    return np.issubdtype(dset.dtype, np.integer) or np.issubdtype(dset.dtype, np.floating)

def squeeze_to_image(array: np.ndarray) -> np.ndarray:
    data = np.array(array)
    while data.ndim > 3:
        data = data[0]
    if data.ndim == 3 and data.shape[-1] == 1:
        data = data[..., 0]
    return data

def normalize_for_display(img: np.ndarray) -> np.ndarray:
    if img.dtype == np.uint8:
        return img
    img = img.astype(np.float32)
    img = img - img.min()
    denom = img.max() if img.max() > 0 else 1.0
    img = img / denom
    return (img * 255).clip(0, 255).astype(np.uint8)

def walk_hdf5(path: Path):
    with h5py.File(path, 'r') as h5:
        summary = []
        images = []
        def visitor(name, obj):
            indent = '  ' * name.count('/')
            if isinstance(obj, h5py.Dataset):
                entry = f"{indent}- dataset: {name} | shape={obj.shape} | dtype={obj.dtype}"
                summary.append(entry)
                if is_image_dataset(obj):
                    images.append(name)
            elif isinstance(obj, h5py.Group):
                entry = f"{indent}- group: {name}"
                summary.append(entry)
        h5.visititems(visitor)
    return summary, images

def preview_images(path: Path, dataset_key: str, max_panels: int = 3):
    with h5py.File(path, 'r') as h5:
        dset = h5[dataset_key]
        if dset.ndim == 3:
            samples = [squeeze_to_image(dset[()])]
            labels = ['all']
        else:
            total = dset.shape[0]
            idxs = np.linspace(0, total - 1, num=min(max_panels, total), dtype=int)
            samples = [squeeze_to_image(dset[idx]) for idx in idxs]
            labels = [str(idx) for idx in idxs]
        fig, axes = plt.subplots(1, len(samples), figsize=(4 * len(samples), 4))
        if len(samples) == 1:
            axes = [axes]
        for ax, sample, label in zip(axes, samples, labels):
            sample = normalize_for_display(sample)
            if sample.ndim == 2:
                ax.imshow(sample, cmap='gray')
            else:
                ax.imshow(sample)
            ax.set_title(f"{dataset_key}\nindex {label}")
            ax.axis('off')
        fig.suptitle(f"{path.name} - {dataset_key}", y=1.02)
        plt.tight_layout()
        plt.show()
```

```python
for path in (human_h5, robot_h5):
    summary, image_like = walk_hdf5(path)
    print(f"\n=== {path.name} ===")
    for line in summary:
        print(line)
    if image_like:
        print(f"\nImage-like datasets detected ({len(image_like)}):")
        for key in image_like:
            print(f"  - {key}")
    else:
        print("\nNo image-like datasets detected")

    if path is human_h5:
        image_like_human = image_like
    else:
        image_like_robot = image_like
```

```python
# Preview the first few image-like datasets (edit the indices as needed)
preview_targets = {
    'human': image_like_human[:2],
    'robot': image_like_robot[:2]
}

for label, keys in preview_targets.items():
    base_path = human_h5 if label == 'human' else robot_h5
    for key in keys:
        preview_images(base_path, key)
```
